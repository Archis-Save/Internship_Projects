#Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#Orders in 2017

```python
df = pd.read_csv("orders_2017.csv")

value = df['total_orders_2017'][0]

plt.figure()
plt.bar(["Orders in 2017"], [value])
plt.ylabel("Total Orders")
plt.title("Total Orders Placed in 2017")
plt.show()
```

#Price vs Purchase Frequency

```python
df = pd.read_csv("price_vs_frequency.csv")

plt.figure()
plt.scatter(df['avg_price'], df['purchase_count'])
plt.xlabel("Average Product Price")
plt.ylabel("Purchase Frequency")
plt.title("Product Price vs Purchase Frequency")
plt.show()
```

#Cumulative Monthly Sales

```python
df = pd.read_csv("cumulative_sales.csv")

plt.figure()
plt.plot(df['month'], df['cumulative_sales'])
plt.xlabel("Month")
plt.ylabel("Cumulative Sales")
plt.title("Cumulative Monthly Sales")
plt.show()
```

#Customer Retention Rate

```python
df = pd.read_csv("customer_retention.csv")

rate = df['retention_rate_percentage'][0]

plt.figure()
plt.bar(["Retention Rate"], [rate])
plt.ylabel("Percentage")
plt.title("Customer Retention Rate (6 Months)")
plt.ylim(0, 5)
plt.show()
```

#Top Customers by Spending

```python
df = pd.read_csv("top_customers.csv")

plt.figure()
plt.bar(df['customer_unique_id'], df['total_spent'])
plt.xlabel("Customer ID")
plt.ylabel("Total Spending")
plt.title("Top Customers by Total Spend")
plt.xticks(rotation=90)
plt.show()
```
